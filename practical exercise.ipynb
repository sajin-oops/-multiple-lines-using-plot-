```python
# - 1. Write a program to drop rows with at least one Nan value (Null value)

import pandas as pd
data = {
    'a': [1,2,None,4],
    'b': [5,6,7,None]
}
df = pd.DataFrame(data)

df_cleaned = df.dropna()
print("Original data dataframe:")
print(df)
print('\n')
print("\nDataFrame after dropping rows with NaN values:")
print(df_cleaned)
```

```python
# - 2.Write a python program to draw multiple lines using plot() function

import matplotlib.pyplot as plt
x = [0,1,2,3,4]
y1 = [0,4,7,9,13]
y2 = [0,5,8,11,22]
plt.plot(x,y1)
plt.plot(x,y2)
plt.title("multiple lines using plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
